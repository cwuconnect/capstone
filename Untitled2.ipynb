```python
import yfinance as yf
import pandas as pd

def get_currency_data(currency, start_date, end_date):
    ticker = f"{currency}=X"  # Append "=X" to the currency code to get the exchange rate
    data = yf.download(ticker, start=start_date, end=end_date)
    return data['Close']

def convert_to_usd(data, exchange_rate):
    return data * exchange_rate

if __name__ == "__main__":
    # Define the currencies and their corresponding exchange rates to USD
    currencies = {
        'EUR': 1.12,
        'GBP': 1.33,
        'JPY': 0.0092,
        'AUD': 0.73,
        'CAD': 0.79,
        'CHF': 1.10,
        'CNY': 0.16,
        'INR': 0.014,
        'BRL': 0.19,
        'MXN': 0.050,
    }

    start_date = '2020-01-01'
    end_date = '2023-01-01'

    # Fetch and convert data for each currency
    for currency, exchange_rate in currencies.items():
        data = get_currency_data(currency, start_date, end_date)
        converted_data = convert_to_usd(data, exchange_rate)

        print(f"\n{currency} Data:")
        print(data.head())
        print(f"\n{currency} Data Converted to USD:")
        print(converted_data.head())
```
